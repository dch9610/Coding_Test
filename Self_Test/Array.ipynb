## 꼭 알아둬야 할 자료구조 : 배열 (Array)

- 데이터를 나열하고, 각 데이터를 인덱스에 대응하도록 구성한 데이터 구조
- 파이썬에서는 리스트 타입이 배열 기능을 제공하고 있음

## 1. 배열이 왜 필요할까?

- 같은 종류의 데이터를 효율적으로 관리하게 위해 사용
- 같은 종류의 데이터를 순차적으로 저장

- 배열의 장점 :
  - 빠른 접근이 가능하다.

- 배열의 단점 :
  - 데이터 추가가 어렵다.
  - 데이터가 삭제되면 뒤에 있는 데이터가 앞으로 당겨진다.

## 2. 파이썬과  C 언어의 배열 예제

- C 언어는 미리 배열 공간을 지정
- 파이썬은 공간을 지정안해도 됨

```python
country ='US'
print(country)
```

```python
# 쉽게 추가가능
country = country + 'A'
print(country)
```
